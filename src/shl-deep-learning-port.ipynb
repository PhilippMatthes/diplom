```python
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('models/shl-model.resnet.h5')
converter = tf.lite.TFLiteConverter.from_keras_model(model)
```

```python
tflite_model = converter.convert()
```

```python
with open('models/shl-model.resnet.tflite', 'wb') as f:
    f.write(tflite_model)
```
